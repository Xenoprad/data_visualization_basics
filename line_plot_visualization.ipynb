```python
import matplotlib.pyplot as plt

x = [1,2,3,4,5] #X-axis data (horizontal)
y = [10,15,13,18,16] # y- axis data (vertical)

plt.figure(figsize=(6,4)) # To set the figure size
plt.plot(x,y, marker = 'o', color='red', label= 'Line plot') # Draw a line
plt.title('Line Plot') #Add title
plt.xlabel ('X-Axis') #Lable X-axis
plt.ylabel ('Y-Axis') #Lable Y-axis
plt.legend()  #Show the label from 'label='
plt.grid(True) #Add grid lines
plt.show() #Display the plot
```
