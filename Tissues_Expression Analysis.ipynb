```python
import matplotlib.pyplot as plt
import pandas as pd
```

```python
Tissues = ['Brain', 'Liver', 'Lung', 'Kidney', 'Heart', 'Pancreas', 'Colon']
Expression_levels = [12.4, 8.7, 15.1, 10.3, 7.2, 5.9, 9.4]
```

```python
plt.bar(Tissues, Expression_levels)
plt.title('Bar chart')
plt.xlabel('Expression_levels') #defining x-axis
plt.ylabel('Tissues')    #defining y-axis
plt.show()
```
